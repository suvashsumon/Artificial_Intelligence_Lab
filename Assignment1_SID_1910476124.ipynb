## Importing necessary modules

```python
import numpy as np
import matplotlib.pyplot as plt
```

## Default Plot Function

```python
def plot_the_equation(x, y, title):
  plt.plot(x, y)
  plt.xlabel('X')
  plt.ylabel('Y')
  plt.title(title)
  plt.show()
```

## Creating X Array

```python
x = np.arange(-10, 10, 0.1)
#print(x)
```

## 1. Function : y = wx + b

```python
w = 5
b = 3
y = w*x + b
#print(y)
plot_the_equation(x, y, "y = wx + b")
```

## 2. Function : y = x^2

```python
y = x**2
plot_the_equation(x, y, "y = x^2")
```

## 3. Function : y = 1/ (1 + e^(-x))

```python
y = 1/ (1 + np.exp(-x))
plot_the_equation(x, y, "y = 1/ (1 + e^(-x))")
```

## 4. Function : y = (e^x - e^(-x)) / (e^x + e^(-x))

```python
y = (np.exp(x) - np.exp(-x)) / (np.exp(x) + np.exp(-x))
plot_the_equation(x, y, "y = (e^x - e^(-x)) / (e^x + e^(-x))")
```

## 5. Function : y = g(f(x)) where u = f(x) = wx + b and y = g(u) = 1/ (1 + e^(-u))

```python
w = 5
b = 3
u = w*x+b
y = 1/(1+np.exp(-u))
plot_the_equation(x, y, "y = g(f(x)) where u = f(x) = wx + b and y = g(u) = 1/ (1 + e^(-u))")
```

## 6. Function : y = g(f(x)) where f(x) = wx + b, u = f(x), and g(u) = (e^u - e^(-u)) / (e^u + e^(-u))

```python
w = 5
b = 3
u = w*x+b
y = (np.exp(u)-np.exp(-u))/(np.exp(u)+np.exp(-u))
plot_the_equation(x, y, "y = g(f(x)) where f(x) = wx + b, u = f(x), and g(u) = (e^u - e^(-u)) / (e^u + e^(-u))")
```

## 7. Function :
y1 = g1(f1(x)) where f1(x) = w1x + b1, u1 = f1(x), and g1(u1) = 1/ (1 + e^(-u1))

y2 = g2(f2(x)) where f2(x) = w2x + b2, u2 = f2(x), and g2(u2) = 1/ (1 + e^(-u2))

y = g3(f3(v)) where f3(v) = w3y1 + w4y2 + b, w = f3(v), and g3(w) = 1/ (1 + e^(-w))

```python
w1 = 7
b1 = 2
u1 = w1*x + b1
y1 = 1/(1+np.exp(-u1))

w2 = 3
b2 = 3
u2 = w2*x + b2
y2 = 1/(1+np.exp(-u2))

w3 = 3.5
w4 = 5.5
b = 4.4
w = w3*y1 + w4*y2 + b
y = 1/(1+np.exp(-w))

plt.plot(x, y, label="y")
plt.plot(x, y1, label="y1")
plt.plot(x, y2, label="y3")
plt.legend()
plt.show()
```
